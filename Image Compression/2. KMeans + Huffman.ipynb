```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from collections import Counter, defaultdict
import heapq
```

```python
# Load image in RGB format
image = cv2.imread('turing.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Reshape the image to a 2D array of pixels (width*height, channels)
pixels = image.reshape(-1, 3)
```

```python
def quantize_colors(pixels, k):
    # Apply K-means clustering to the pixel data
    kmeans = KMeans(n_clusters=k, random_state=0).fit(pixels)
    
    # Replace each pixel's RGB value with the centroid of its cluster
    quantized_pixels = kmeans.cluster_centers_[kmeans.labels_]
    quantized_pixels = np.round(quantized_pixels).astype(int)
    
    # Reshape back to the original image dimensions
    quantized_image = quantized_pixels.reshape(image.shape)
    return quantized_image, kmeans

# Define number of clusters (tune for compression-quality trade-off)
k = 16
quantized_image, kmeans = quantize_colors(pixels, k)
```

```python
# Map each pixel to its cluster index
cluster_labels = kmeans.labels_

# Count the frequency of each cluster label (color) in the quantized image
cluster_freq = Counter(cluster_labels)
```

```python
class HuffmanNode:
    def __init__(self, color, freq):
        self.color = color
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(freq_dict):
    heap = [HuffmanNode(color, freq) for color, freq in freq_dict.items()]
    heapq.heapify(heap)
    
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = HuffmanNode(None, left.freq + right.freq)
        merged.left = left
        merged.right = right
        heapq.heappush(heap, merged)
    
    return heap[0]

def generate_huffman_codes(node, prefix='', codebook={}):
    if node is not None:
        if node.color is not None:
            codebook[node.color] = prefix
        generate_huffman_codes(node.left, prefix + '0', codebook)
        generate_huffman_codes(node.right, prefix + '1', codebook)
    return codebook

# Build the Huffman tree and generate codes
huffman_tree_root = build_huffman_tree(cluster_freq)
huffman_codes = generate_huffman_codes(huffman_tree_root)
```

```python
# Flatten the cluster labels for easy encoding
encoded_image = ''.join(huffman_codes[label] for label in cluster_labels)
```

```python
import pickle

# Metadata includes the cluster centers and Huffman codes
metadata = {
    'cluster_centers': kmeans.cluster_centers_,
    'huffman_codes': huffman_codes
}

# Save metadata and compressed image
with open('compressed_image_data.pkl', 'wb') as f:
    pickle.dump((metadata, encoded_image), f)
```

```python
def decode_huffman(encoded_image, huffman_codes):
    reverse_huffman_codes = {v: k for k, v in huffman_codes.items()}
    decoded_labels = []
    code = ''
    for bit in encoded_image:
        code += bit
        if code in reverse_huffman_codes:
            decoded_labels.append(reverse_huffman_codes[code])
            code = ''
    return decoded_labels

def decompress_image(encoded_image, metadata):
    # Decode the Huffman-encoded data
    decoded_labels = decode_huffman(encoded_image, metadata['huffman_codes'])
    
    # Map decoded labels back to cluster center RGB values
    cluster_centers = metadata['cluster_centers']
    decompressed_pixels = cluster_centers[decoded_labels].astype(np.uint8)
    
    # Reshape to the original image dimensions
    height, width, _ = image.shape
    decompressed_image = decompressed_pixels.reshape((height, width, 3))
    return decompressed_image

# Load metadata and compressed image
with open('compressed_image_data.pkl', 'rb') as f:
    metadata, encoded_image = pickle.load(f)

# Decompress
decompressed_image = decompress_image(encoded_image, metadata)
```

```python
# Plot the original, quantized, and decompressed images
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(quantized_image.astype(np.uint8))
plt.title(f'Quantized Image (k={k})')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(decompressed_image)
plt.title('Decompressed Image')
plt.axis('off')

plt.show()
```
