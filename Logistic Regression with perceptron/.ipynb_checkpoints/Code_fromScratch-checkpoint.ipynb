<h2> Perceptron from scratch </h2>

<h3> Activation function: Heaviside step function </h3>

```python
# Generating synthetic datasets:
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
# Perceptron code:

def step(z):
    return 1 if z>0 else 0

def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0], weights[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

<h3> Activation function: Sigmoid </h3>

```python
# Sigmoid Function:
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
    
z_values = np.linspace(-10, 10, 100)
sigmoid_values = sigmoid(z_values)

plt.figure(figsize=(8, 6))
plt.plot(z_values, sigmoid_values, label='Sigmoid Function', color='b', linewidth=2)
plt.title('Sigmoid Function')
plt.xlabel('Input (z)')
plt.ylabel('Output (Sigmoid(z))')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axhline(1, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
# The derivative of the sigmoid function

def sigmoid_derivative(z):
    sig = sigmoid(z)
    return sig * (1 - sig)

sigmoid_derivative_values = sigmoid_derivative(z_values)

plt.figure(figsize=(8, 6))
plt.plot(z_values, sigmoid_values, label='Sigmoid Function', color='b', linewidth=2)
plt.plot(z_values, sigmoid_derivative_values, label='Sigmoid Derivative', color='r', linestyle='--', linewidth=2)
plt.title('Sigmoid Function and its Derivative')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axhline(1, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
# Perceptron code:

def sigmoid(z):
    return (1/(1+np.exp(-z)))
    
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m_logistic = -(lor.coef_[0][0] / lor.coef_[0][1])  
b_logistic = -(lor.intercept_ / lor.coef_[0][1])   

x_input1 = np.linspace(-3, 3, 100) 
y_input1 = m_logistic * x_input1 + b_logistic  

m_perceptron = -coef_[0] / coef_[1]  
b_perceptron = -intercept_ / coef_[1]  

x_input_perceptron = np.linspace(-3, 3, 100)
y_input_perceptron = m_perceptron * x_input_perceptron + b_perceptron

# Plotting
plt.figure(figsize=(10, 6))

# Plot perceptron decision boundary (red line)
plt.plot(x_input_perceptron, y_input_perceptron, color='red', linewidth=3, label='Perceptron')

# Plot logistic regression decision boundary (black line)
plt.plot(x_input1, y_input1, color='black', linewidth=3, label='Logistic Regression')

# Scatter plot of the data points
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)

# Set limits for y-axis
plt.ylim(-3, 2)

# Add labels and legend
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# Show the plot
plt.show()
```

<h3> Perceptron with Sigmoid as Activation Function and apply Gradient Descent: </h3>

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty=None,solver='sag')
lor.fit(X,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
# Our function:

def sigmoid(z):
    return 1/(1 + np.exp(-z))
    
def perceptron_logistic(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    
    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
        
    return weights[1:],weights[0]
```

```python
coef_,intercept_ = perceptron_logistic(X,y)
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
